```python
from bs4 import BeautifulSoup
import requests
from openpyxl import Workbook
import pandas as pd
website='https://www.coingecko.com/en'
response=requests.get(website)
response.status_code
#pada baris ini kita mengimport libary dan menginisliasi sekaligus request terhadap web yang akan discrap
```

```python
soup=BeautifulSoup(response.content,'html.parser') 
results=soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
(results)
#Pada baris ini kita mengambil parse html dari web yang discrap
```

```python
#Melakukan inisiliasi variabel
name=[]
price=[]
change1h=[]
change24h=[]
change7d=[]
volume24h=[]
marketcap=[]
#Melakukan scraping semua page yang ada pada coingecko
for page in range (1,138):
    print(page)
    website='https://www.coingecko.com/en?page='+str(page)
    response = requests.get(website)
    soup=BeautifulSoup(response.content,'html.parser')
    results=soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
    for result in results:
#memberikan conditional pada pencarian data yang akan discrap agar mudah menghandle error 
        try:
            name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
        except:
            name.append('NA')
        try:
            price.append(result.find('td', {'class':'td-price price text-right pl-0'}).get_text().strip())
        except:
            price.append('NA')
        try:
            change1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
        except:
            change1h.append('NA')
        try:
            change24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
        except:
            change24h.append('NA')
        try:
            change7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
        except:
            change7d.append('NA')
        try:
        #Dibagian volume ini saya bingung sudah ambil classnya bahkan semua class yang ada pada table volume tapi tetap error
            volume24h.append(result.results[0].find('td', {'class':'td-liquidity_score lit text-right col-market'}).get_text().strip())
        except:
            volume24h.append('NA')
        try:
            marketcap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
        except:
            marketcap.append('NA')   
        #Pada bagian ini kita akan melakukan scraping dari page 1-137 
```

```python
tabeldata=pd.DataFrame({'Token':name, 'Harga':price, '1 jam':change1h,'24 Jam':change24h,'1 Minggu':change7d,'Volume':volume24h,'MarketCap':marketcap})
tabeldata
#membuat tabel agar mudah dibaca 
```

```python
#Menjadikan data sebagai dataset dalam bentuk xlsx 
exportdata=tabeldata.to_excel('datascraping.xlsx', index=False)
```
